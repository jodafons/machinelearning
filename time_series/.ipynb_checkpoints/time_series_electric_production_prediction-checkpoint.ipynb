## Time Series Studies Using Machine Learning

```python
import pandas as pd
from keras import *
import numpy as np
import matplotlib.pyplot as plt 
import warnings
from sklearn.metrics import *
import statsmodels.formula.api as smf # statistics and econometrics
import statsmodels.tsa.api as smt
import statsmodels.api as sm
import scipy.stats as scs
warnings.filterwarnings('ignore')

%config InlineBackend.figure_format = 'retina'

def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

Reading the dataset using pandas library.

```python
data = pd.read_csv('samples/Electric_Production.csv', 
                   parse_dates=['DATE'],
                   index_col=['DATE'])
print(data.head())
#print(data.describe())
series = data.IPG2211A2N
plt.figure(figsize=(15, 7))
plt.plot(series, 'k')
plt.title('Electrical power (1984 to 2018)')
plt.grid(True)
plt.show()
```

## Move, smoothe, evaluate

Moving average (smoothe) function:

```python
def plot_moving_average(series, window, 
                        plot_intervals=False, 
                        scale=1.96, 
                        plot_anomalies=False):

    """
        series - dataframe with timeseries
        window - rolling window size 
        plot_intervals - show confidence intervals
        plot_anomalies - show anomalies 

    """
    rolling_mean = series.rolling(window=window).mean()

    plt.figure(figsize=(15,5))
    plt.title("Moving average\n window size = {}".format(window))
    plt.plot(rolling_mean, "g", label="Rolling mean trend")

    # Plot confidence intervals for smoothed values
    if plot_intervals:
        mae = mean_absolute_error(series[window:], rolling_mean[window:])
        deviation = np.std(series[window:] - rolling_mean[window:])
        lower_bond = rolling_mean - (mae + scale * deviation)
        upper_bond = rolling_mean + (mae + scale * deviation)
        plt.plot(upper_bond, "r--", label="Upper Bond")
        plt.plot(lower_bond, "r--", label='Lower Boud')
        
        # Having the intervals, find abnormal values
        #if plot_anomalies:
        #    anomalies = pd.DataFrame(index=series.index, columns=series.columns)
        #    anomalies[series<lower_bond] = series[series<lower_bond]
        #    anomalies[series>upper_bond] = series[series>upper_bond]
        #    plt.plot(anomalies, "ro", markersize=10)
        
    plt.plot(series[window:], label="Actual values")
    plt.legend(loc="upper left")
    plt.grid(True)
```

### 3 months of lag (window)

```python
# 3 months lag
plot_moving_average(series,3,plot_intervals=True, plot_anomalies=False)
```

### 5 mounths of lag (window)

```python
plot_moving_average(series,5,plot_intervals=True, plot_anomalies=False)
```

Checking the autocorrelation. Let's pain attention in the Dickey-Fuller test.


```python
def tsplot(y, lags=None, figsize=(12, 7), style='bmh'):
    """
        Plot time series, its ACF and PACF, calculate DickeyFuller test
        y - timeseries
        lags - how many lags to include in ACF, PACF calculation
    """
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
        
    with plt.style.context(style):    
        fig = plt.figure(figsize=figsize)
        layout = (2, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))
        y.plot(ax=ts_ax)
        p_value = sm.tsa.stattools.adfuller(y)[1]
        ts_ax.set_title('Time Series Analysis Plots\n Dickey-Fuller: p={0:.5f}'.format(p_value))
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax)
        plt.tight_layout()
        
        
tsplot( series - result_add.trend - result_add.seasonal, 100)
```

## Create the cross validation method

For time series, we will use the TimeSeriesSplit from sklearn.

```python
def timeseries_train_test_split(X, y, test_size):
    # Perform train-test split with respect to time series structure
    # get the index after which test set starts
    test_index = int(len(X)*(1-test_size))
    X_train = X.iloc[:test_index]
    y_train = y.iloc[:test_index]
    X_test = X.iloc[test_index:]
    y_test = y.iloc[test_index:]
    return X_train, X_test, y_train, y_test


from sklearn.linear_model import LinearRegression
# for time-series cross-validation set 5 folds 
tscv = TimeSeriesSplit(n_splits=5)
```

## Create some features

```python
# Lets add some lags between 1 and 10. 11 features
d = pd.DataFrame(data.copy())
d.columns = ['y']
for i in range(1, 15+1):
    d["lag_{}".format(i)] = d.y.shift(i)
print(d.head())
```

## Create the model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

# remove NaN values and use the first column as target
y = d.dropna().y

# Use all columns as Train/Test. (Remove NaN values)
X = d.dropna().drop(['y'], axis=1)

# Create the train and test dataset
X_train, X_test, y_train, y_test = timeseries_train_test_split(X, y, test_size=0.3)

# machine learning in two lines
lr = LinearRegression()
lr.fit(X_train, y_train)
```

## Plot the test and check the Mean Absolute Percentage Error

The test set will be the novelty. The red lines is given by the upper (mae+std) and lower (mae-std) boud caluculated from the validation set in the cross validation method. The test set can not violate these lines since your prediction must be inside of these bouds.

```python
def plot_model_results(model, X_train, X_test, 
                       plot_intervals=False, plot_anomalies=False):
    
    # Plots modelled vs fact values, prediction intervals and anomalies
    prediction = model.predict(X_test)
    
    plt.figure(figsize=(15, 7))
    plt.plot(prediction, "g", label="prediction", linewidth=2.0)
    plt.plot(y_test.values, label="actual", linewidth=2.0)
    
    if plot_intervals:
        cv = cross_val_score(model, X_train, y_train, 
                                    cv=tscv, 
                                    scoring="neg_mean_absolute_error")
        mae = cv.mean() * (-1)
        deviation = cv.std()
        
        scale = 1.96
        lower = prediction - (mae + scale * deviation)
        upper = prediction + (mae + scale * deviation)
        
        plt.plot(lower, "r--", label="upper bond / lower bond", alpha=0.5)
        plt.plot(upper, "r--", alpha=0.5)
        
        if plot_anomalies:
            anomalies = np.array([np.NaN]*len(y_test))
            anomalies[y_test<lower] = y_test[y_test<lower]
            anomalies[y_test>upper] = y_test[y_test>upper]
            plt.plot(anomalies, "o", markersize=10, label = "Anomalies")
    
    error = mean_absolute_percentage_error(prediction, y_test)
    plt.title("Mean absolute percentage error {0:.2f}%".format(error))
    plt.legend(loc="best")
    plt.tight_layout()
    plt.grid(True);
    
plot_model_results(lr, X_train, X_test, plot_intervals=True, plot_anomalies=True)
```

## Use MLP model

Lets try to use the MLP model with lags in the input. Here, the feature extraction is did by me using lag box.

### Create some features:

```python
# Lets add some lags between 1 and 10. 
# We have 10 features
d = pd.DataFrame(data.copy())
d.columns = ['y']
for i in range(1, 10+1):
    d["lag_{}".format(i)] = d.y.shift(i)

# remove NaN values and use the first column as target
y = d.dropna().y

# Use all columns as Train/Test. (Remove NaN values)
X = d.dropna().drop(['y'], axis=1)

# Create the train and test dataset (10 features)
X_train, X_test, y_train, y_test = timeseries_train_test_split(X, y, test_size=0.3)

print(X_train.head())
```

### Create and Fit the MLP model

```python
from keras.models import Sequential
from keras.layers import Dense
import keras.backend as K
from keras.callbacks import EarlyStopping

X_train_np = X_train.values
X_test_np = X_test.values
y_train_np = y_train.values
y_test_np = y_test.values

# MLP simple model just for test case
model = Sequential()
model.add(Dense(10, input_dim=10, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
early_stop = EarlyStopping(monitor='loss', patience=2, verbose=1)

# Before we start, we need to do some normalization because of large values in the 
# dataset (X).
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_train_np_norm = sc.fit_transform(X_train_np)
X_test_np_norm = sc.transform(X_test_np)
history = model.fit(X_train_np_norm, y_train_np, epochs=200, batch_size=1, 
                    verbose=0, callbacks=[early_stop], shuffle=False)
```

### Evaluate the model using the test dataset


```python

def plot_model_results(model, X_test, y_test):  
    # Plots modelled vs fact values, prediction intervals and anomalies
    prediction = model.predict(X_test)
    plt.figure(figsize=(15, 7))
    plt.plot(prediction, "g", label="prediction", linewidth=2.0)
    plt.plot(y_test, label="actual", linewidth=2.0)
    error = mean_absolute_percentage_error(prediction, y_test)
    plt.title("Mean absolute percentage error {0:.2f}%".format(error))
    plt.legend(loc="best")
    plt.tight_layout()
    plt.grid(True);
    
plot_model_results(model, X_test_np_norm, y_test_np)
```
