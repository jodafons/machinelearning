# Time Series USD Prediction Using Machine Learning

```python
# Import evething here

import pandas as pd
import numpy as np
import keras.backend as K
import matplotlib.pyplot as plt 
import warnings
import statsmodels.formula.api as smf # statistics and econometrics
import statsmodels.tsa.api as smt
import statsmodels.api as sm
import scipy.stats as scs
from dateutil.parser import parse
from sklearn.linear_model import LinearRegression
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import *
from keras.callbacks import EarlyStopping,ModelCheckpoint
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.preprocessing import MinMaxScaler


warnings.filterwarnings('ignore')
%config InlineBackend.figure_format = 'retina'

def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Use this to hold the information for each strategy
summary = {}
```

### Set the dataset USD from the sample files

```python
data = pd.read_csv('samples/USD_INR.csv', index_col=['Date'], parse_dates=['Date'])
#data["Date"] = pd.to_datetime(data["Date"])
#data = data.set_index(["Date"], drop=True)
#data = data.sort_index(axis=1 ,ascending=True)
data = data.iloc[::-1]
#print(data.head())

series = data.Price
print(series.head())
plt.figure(figsize=(15, 7))
plt.plot(series, 'k')
plt.title('USD INR')
plt.grid(True)
plt.show()
```

### Mode Average Studies

```python
def plot_moving_average(series, window, 
                        plot_intervals=False, 
                        scale=1.96, 
                        ):

    rolling_mean = series.rolling(window=window).mean()
    plt.figure(figsize=(15,5))
    plt.title("Moving average\n window size = {}".format(window))
    plt.plot(rolling_mean, "g", label="Rolling mean trend")

    # Plot confidence intervals for smoothed values
    if plot_intervals:
        mae = mean_absolute_error(series[window:], rolling_mean[window:])
        deviation = np.std(series[window:] - rolling_mean[window:])
        lower_bond = rolling_mean - (mae + scale * deviation)
        upper_bond = rolling_mean + (mae + scale * deviation)
        plt.plot(upper_bond, "r--", label="Upper Bond")
        plt.plot(lower_bond, "r--", label='Lower Boud')
        
    plt.plot(series[window:], label="Actual values")
    plt.legend(loc="upper left")
    plt.grid(True)
    
    
plot_moving_average(series, 4, plot_intervals=True)
plot_moving_average(series, 40, plot_intervals=True)
plot_moving_average(series, 400, plot_intervals=True)
```

Moving average strategy can be a good trend extimation approch. Lets keep this idea.

### Series Decomposition Studies

```python
def plot_decomposition(res):
    fig, axes = plt.subplots(ncols=1, nrows=4, sharex=True, figsize=(12,5))
    res.observed.plot(ax=axes[0], legend=False)
    axes[0].set_ylabel('Observed')
    res.trend.plot(ax=axes[1], legend=False)
    axes[1].set_ylabel('Trend')
    res.seasonal.plot(ax=axes[2], legend=False)
    axes[2].set_ylabel('Seasonal')
    res.resid.plot(ax=axes[3], legend=False)
    axes[3].set_ylabel('Residual')
    plt.tight_layout()
    plt.show()



# Add Decomposition: Observed = Trend + Seasonal + Residual 
result_add = seasonal_decompose(series, model='additive', extrapolate_trend='freq',freq=100)
plot_decomposition(result_add)
```

```python
# Mul Decomposition: Observed = Trend * Seasonal * Residual 
result_add = seasonal_decompose(series, model='mul', extrapolate_trend='freq',freq=100)
plot_decomposition(result_add)
```

### Check the Auto Correlation

```python
def tsplot(y, lags=None, figsize=(12, 7), style='bmh'):
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):    
        fig = plt.figure(figsize=figsize)
        layout = (2, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))
        y.plot(ax=ts_ax)
        p_value = sm.tsa.stattools.adfuller(y)[1]
        ts_ax.set_title('Time Series Analysis Plots\n Dickey-Fuller: p={0:.5f}'.format(p_value))
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax)
        plt.tight_layout()
        
tsplot(series, 10)
```

### Create the cross validation 

```python
# for time-series cross-validation set 10 folds (Train/Val)
from sklearn.model_selection import TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=10)
```

# Use Linear Regression Model to Predict

For the first approch we will use a linear model and fead this model with lags patterns.

```python
# Train (Train+Val) and Test
def timeseries_train_test_split(X, y, test_size):
    # Perform train-test split with respect to time series structure
    # get the index after which test set start
    test_index = int(len(X)*(1-test_size))
    X_train = X.iloc[:test_index]
    y_train = y.iloc[:test_index]
    X_test = X.iloc[test_index:]
    y_test = y.iloc[test_index:]
    return X_train, X_test, y_train, y_test


# for time-series cross-validation set 10 folds (Train/Val)
from sklearn.model_selection import TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=10)


def cross_validation_fit(X,y,cv):
    from sklearn.metrics import mean_absolute_error
    from sklearn.linear_model import LinearRegression
    mae = []; models = []
    for train_index, val_index in cv.split(X):
        # Get the train and val
        X_trn      = X.iloc[train_index].values
        y_true_trn = y.iloc[train_index].values
        X_val        = X.iloc[val_index].values
        y_true_val   = y.iloc[val_index].values
        # create the linear regression model
        model = LinearRegression()
        # fit the model
        model.fit(X_trn,y_true_trn)
        # predict the validation
        y_pred_val = model.predict(X_val)
        mae.append( mean_absolute_error(y_true_val, y_pred_val) )
        models.append(model)
    return model, mae
```

### Create all patterns

```python
d = pd.DataFrame(series.copy())
d.columns = ['y']
for i in range(1, 10):
    d["lag_{}".format(i)] = d.y.shift(i)
    
# remove NaN values and use the first column as target
y = d.dropna().y
# Use all columns as Train/Test. (Remove NaN values)
X = d.dropna().drop(['y'], axis=1)
```

### Calculate the fluctuation

The point has some fluctuation like point+error+deviation for both sides (upper or lower bond). The cross validation will estimate these values using the validation dataset to estimate and train dataset to fit the model. The operation model will be training using train and validation dataset. The test dataset will be used to test the model.

```python
X_train, X_test, y_train, y_test = timeseries_train_test_split(X, y, test_size=0.1)

# Calculate the fluctuation 
models, mae = cross_validation_fit(X_train, y_train, tscv)
# get mean and standard desviation values
mae_mean = np.mean(mae); mae_std = np.std(mae)
print('MAE: ', mae_mean,'±',mae_std)
```

### Calculate the operation model

```python
# Get the operation model
model = LinearRegression()
# fit the model
model.fit(X_train,y_train)
```

### Evaluate the model

```python
def plot_model_results(y_pred, y_true, 
                       mean_value_from_cv_score=None, 
                       std_value_from_cv_score=None):
    
    plt.figure(figsize=(15, 7))
    plt.plot(y_pred, "g", label="prediction", linewidth=2.0)
    plt.plot(y_true, label="actual", linewidth=2.0)
    if mean_value_from_cv_score and std_value_from_cv_score:
        scale = 1.96; m=mean_value_from_cv_score; std=std_value_from_cv_score
        lower = y_pred - (m + scale * std)
        upper = y_pred + (m + scale * std)    
        plt.plot(lower, "r--", label="upper bond / lower bond", alpha=0.5)
        plt.plot(upper, "r--", alpha=0.5)
    #anomalies = np.array([np.NaN]*len(y_test))
    #anomalies[y_test<lower] = y_test[y_test<lower]
    #anomalies[y_test>upper] = y_test[y_test>upper]
    #plt.plot(anomalies, "o", markersize=10, label = "Anomalies")
    error = mean_absolute_percentage_error(y_pred, y_true)
    plt.title("Mean absolute percentage error {0:.2f}%".format(error))
    plt.legend(loc="best")
    plt.tight_layout()
    plt.grid(True);
    
y_pred_test = model.predict(X_test)
plot_model_results(y_pred_test, y_test.values,mae_mean, mae_std)

from copy import deepcopy
summary['linear'] = {'model': deepcopy(model), 'mae_cv_score':mae }
```

# Use MLP Model to Predict

Prepare the Data to the training phase:

```python
# test size percentage
test_size = 0.2
test_index = int(len(data)*(1-test_size))

# Normalization
d = pd.DataFrame(series.copy())
d.columns = ['y']
sc = MinMaxScaler()
d.y = np.concatenate((sc.fit_transform(d.iloc[:test_index]), 
                    sc.transform(d.iloc[test_index:])))

# Create features
for lag in range(1,10):
    d["lag_%d"%lag] = d.y.shift(lag)
# remove NaN values and use the first column as target
y = d.dropna().y
# Use all columns as Train/Test. (Remove NaN values)
X = d.dropna().drop(['y'], axis=1)

# Split: train and test
X_train, X_test, y_train, y_test = timeseries_train_test_split(X, y, test_size=0.1)
X_train = X_train.values;
X_test  = X_test.values
y_train = y_train.values
y_test  = y_test.values
```

### Create the MLP and Train

```python
K.clear_session()
model_mlp = Sequential()
model_mlp.add(Dense(12, input_dim=X_train.shape[1], activation='relu'))
model_mlp.add(Dense(1))
```

### Calculate the Fluctuation

```python
def cross_validation_fit(model,X,y,cv):
    from sklearn.metrics import mean_absolute_error
    from keras.models import clone_model
    mae = []; models = []; history_list = []
    for cv_i, index in enumerate(cv.split(X)):
        
        train_index = index[0]
        val_index = index[1]
        # Get the train and val
        X_trn        = X[train_index]
        y_true_trn   = y[train_index]
        X_val        = X[val_index]
        y_true_val   = y[val_index]
        my_model = clone_model(model)
        # fit the model
        filepath="weights.best.hdf5"
        print('Training the model using the fold: %d' % cv_i)
        my_model.compile(loss='mean_squared_error', optimizer='adam')
        early_stop = EarlyStopping(monitor='val_loss', patience=10, verbose=0)
        checkpoint = ModelCheckpoint(filepath, monitor='val_loss', 
                                     verbose=0, 
                                     save_best_only=True, 
                                     mode='min')
        history = my_model.fit(X_trn, y_true_trn, epochs=10, 
                                    batch_size=1, 
                                    verbose=0, 
                                    shuffle=False, 
                                    validation_data = (X_val, y_true_val),
                                    callbacks=[early_stop, checkpoint])
        my_model.load_weights(filepath)
        # predict the validation
        y_pred_val = my_model.predict(X_val)
        mae.append( mean_absolute_error(y_true_val, y_pred_val) )
        models.append(my_model)
        history_list.append( history )
    return models, history_list, mae 


print(X_train.shape)
print(y_train.shape)
# Calculate the fluctuation 
models, history, mae = cross_validation_fit(model_mlp, X_train, y_train, tscv)
```

### Train the Operation Model

```python
filepath="weights.best.hdf5"
K.clear_session()
model_mlp = Sequential()
model_mlp.add(Dense(12, input_dim=X_train.shape[1], activation='relu'))
model_mlp.add(Dense(1))
model_mlp.compile(loss='mean_squared_error', optimizer='adam')
early_stop = EarlyStopping(monitor='loss', patience=10, verbose=0)
checkpoint = ModelCheckpoint(filepath, monitor='loss', 
                                     verbose=0, 
                                     save_best_only=True, 
                                     mode='min')
history = model_mlp.fit(X_train, y_train, epochs=10, 
                                    batch_size=1, 
                                    verbose=0, 
                                    shuffle=False, 
                                    callbacks=[early_stop, checkpoint])
model_mlp.load_weights(filepath)
# predict the validation
y_pred_test = model_mlp.predict(X_test)
```

### Evaluate the Model

```python
# We must inverter the normalization for an real estimation.
y_pred_test = model_mlp.predict(X_test)

plot_model_results(y_pred_test, y_test.reshape((y_test.shape[0],1)),
                   np.mean(mae), np.std(mae))


print('After inverse tranformation')
plot_model_results(sc.inverse_transform(y_pred_test), 
                   sc.inverse_transform(y_test.reshape((y_test.shape[0],1))),
                   sc.inverse_transform([[np.mean(mae)]]), 
                   sc.inverse_transform([[np.std(mae)]]))


summary['mlp'] = {'model': model_mlp, 'mae_cv_score':mae, 
                  'history':history , 'cv_models':models}
```
